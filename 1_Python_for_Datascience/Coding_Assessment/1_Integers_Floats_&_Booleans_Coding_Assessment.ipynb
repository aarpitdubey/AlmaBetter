## Mid point of a line

### Ques: Write a Python program to calculate midpoints of a line.



```python
# Step 1 : Creating a function which will calculate the mid points

def midpoint(x1, y1, x2, y2):
    # Step 2 :  mid_point formula
    midp_x = (x1 + x2) / 2
    midp_y = (y1 + y2) / 2
        
    return str(midp_x)+" "+str(midp_y)

print("*"*70)
print(" "*30 + "INPUT" + " "*30)
print("*"*70)

# Inputs for coordinates
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

print("*"*70)
print(" "*30 + "RESULT" + " "*30)
print("*"*70)
print(midpoint(x1, y1, x2, y2))

```
